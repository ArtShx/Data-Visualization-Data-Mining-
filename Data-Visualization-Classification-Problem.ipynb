```python
'''
    Tutorial website: https://www.springboard.com/blog/data-mining-python-tutorial/
'''
import pandas
import matplotlib
import matplotlib.pyplot
import numpy

import sklearn
from sklearn import cluster

%matplotlib inline

faithful = pandas.read_csv(r'C:\Users\silvaart\Desktop\workspace\pessoal\Programacao\AI-playing-Pong-master\Data Visualization - Data Mining\classification\faithful_2.csv')
faithful.head()
```

```python
faithful.columns = ['eruptions', 'waiting']

matplotlib.pyplot.scatter(faithful.eruptions, faithful.waiting)
matplotlib.pyplot.title('Old Faithful Data Scatterplot')
matplotlib.pyplot.xlabel('Length of Eruptions (minutes)')
matplotlib.pyplot.ylabel('Time between eruptions (minutes)')
```

```python
# setting up the variables for creating a cluster model
faith = numpy.array(faithful) # read csv as numpy array in order for sci-kit be able to read the data
k = 2 # number of clusters (categories)
kmeans = cluster.KMeans(n_clusters=k)
kmeans.fit(faith)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# creating the visualization
for i in range(k):
    # select only data observations with cluster label == i
    ds = faith[numpy.where(labels==i)]
    # plot the data observations
    matplotlib.pyplot.plot(ds[:,0], ds[:,1], 'o', markersize=7)
    # plot the centroids
    lines = matplotlib.pyplot.plot(centroids[i, 0], centroids[i, 1], 'kx')
    # make the centroid x's bigger
    matplotlib.pyplot.setp(lines, ms=15.0)
    matplotlib.pyplot.setp(lines, mew=4.0)
matplotlib.pyplot.show()
```
